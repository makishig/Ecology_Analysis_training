##GLM(一般化線形モデル)
今までの相関は２変数の関係しか見れず、他の要因を無視している
GLMでは複数の環境要因を同時に扱える
ex)  個体数=光+水分+温度+誤差
それぞれの効果を分離することができ、各要因の効果の大きさが分かる

```python
#簡単な例
#サンプルデータ
import numpy as np
import pandas as pd

np.random.seed(0)

n = 100 #サンプル数
light = np.random.uniform(0, 10, n)
moisture = np.random.uniform(0, 5, n)

#真の関係
count = 2*light +3*moisture + np.random.normal(0, 2, n)

df = pd.DataFrame({
    'light' : light,
    'moisture' : moisture,
    'count' : count
})

```

```python
df.corr()
```

```python
import statsmodels.api as sm

x = df[['light', 'moisture']]
x = sm.add_constant(x)
y = df['count']

model = sm.GLM(y, x, family=sm.families.Gaussian())
result = model.fit()   #モデルの実行

print(result.summary())
```
