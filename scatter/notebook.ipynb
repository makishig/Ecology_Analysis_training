#データの可視化と基本統計
##サンプルデータを用い、基本統計量の計算と散布図の作成

```python
import numpy as np
import matplotlib.pyplot as plt
#再現性のためシードの固定
rng=np.random.default_rng(123)
#サンプルデータ作成
x=rng.normal(0,1,50)
y=2*x+rng.normal(0,0.5,50)
print('xの平均',np.mean(x))
print('xの分散',np.var(x))
print('yの平均',np.mean(y))
print('yの分散',np.var(y))
#可視化
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('scatter plot of x and y')
plt.show()
```
