#ピアソン相関
ピアソンは値そのものの線形関係を見る解析
直線関係を測るものであるため、曲線（山形）などでは相関が弱くなる
また元の値をそのまま使う、外れ値に弱い特徴

```python
import numpy as np
import pandas as pd

np.random.seed(0)
n=50 #サンプル数５０

light = np.random.uniform(0,10,n) #0~10のランダム、サンプル数n
biomass = 2* light + np.random.normal(0,2,n) #平均０分散２のノイズを光量に足す

df = pd.DataFrame({
    'light':light,
    'biomass':biomass
})
df.head()
```

```python
#ピアソン相関
from scipy.stats import pearsonr

r,p = pearsonr(df['light'],df['biomass'])
print(r,p)
```

r:関係の強さと向き　値が大きければ強い相関がある
p:その関係が偶然かどうか 小さいほど有意と言える 統計的な確からしさ

ex)
r = 0.8, p = 0.2   相関があるように見えるが偶然の可能性が高い　データ不足
r = 0.2, p < 0.01  関係は弱いが有意　データが多い

なぜサンプル数でp値が変わるのか
データが多いほど、小さな相関でも偶然ではないと判断されやすくなるため

```python
#可視化
import matplotlib.pyplot as plt
plt.scatter(df['light'],df['biomass'])
plt.xlabel('light')
plt.ylabel('biomass')
plt.show()
```
