##GLMM
#GLM + ランダム効果
#GLMはすべてのデータが独立と仮定している。しかし調査では同じ調査地で複数サンプル(10 site * 10 plot　みたいに)、年ごとのデータが存在するなど、データが独立ではない。     
モデルの形  個体数=光+水分+サイト差+誤差

固定効果：知りたい効果（光、水分）、ランダム効果：ばらつきとして扱う（サイト差、個体差）

GLMMでサイト差を考慮し、場所の違いを取り除ける

```python
#サンプルデータ 3site * 3plot 個体数、光、水分
import numpy as np
import pandas as pd
data = {
    "site": ["A","A","A","B","B","B","C","C","C"],
    "light": [5.2, 4.8, 6.1, 3.2, 3.8, 3.5, 7.0, 6.5, 7.2],
    "moisture": [2.1, 2.3, 2.0, 3.5, 3.8, 3.2, 1.5, 1.8, 1.6],
    "count": [10, 12, 9, 5, 6, 4, 15, 14, 16]
}
df = pd.DataFrame(data)
df
```

```python
#GLMMの実行
import statsmodels.formula.api as smf

model = smf.mixedlm("count ~ light + moisture", df, groups=df["site"]) #おそらく、count = a*light + b*moisture の関係を示してる
result = model.fit()

print(result.summary())
```

siteが少なすぎてsite間のランダム効果が求められなかった
より大きなデータでもう一回

```python
 #サンプルデータ
#A：中間環境
#B：湿潤・低光 → 個体数少
#C：高光・乾燥 → 個体数多
#D：低光・高水分 → 個体数少
#E：中〜高光 → 中程度
#サイトごとに違いがあるように

data = {   
    "site": [
        "A","A","A","A","A",
        "B","B","B","B","B",
        "C","C","C","C","C",
        "D","D","D","D","D",
        "E","E","E","E","E"
    ],
    
    "light": [
        5.2, 4.8, 6.1, 5.5, 6.3,
        3.2, 3.8, 3.5, 4.0, 3.6,
        7.0, 6.5, 7.2, 6.8, 7.5,
        2.5, 2.8, 3.0, 2.7, 3.2,
        6.0, 6.3, 5.8, 6.5, 6.1
    ],
    
    "moisture": [
        2.1, 2.3, 2.0, 2.2, 1.9,
        3.5, 3.8, 3.2, 3.6, 3.4,
        1.5, 1.8, 1.6, 1.7, 1.4,
        4.0, 4.2, 3.8, 4.1, 3.9,
        2.5, 2.3, 2.7, 2.4, 2.6
    ],
    
    "count": [
        10, 12, 9, 11, 13,
        5, 6, 4, 7, 5,
        15, 14, 16, 15, 17,
        3, 4, 5, 3, 4,
        11, 12, 10, 13, 12
    ]
}
```

```python
df = pd.DataFrame(data)

model = smf.mixedlm('count ~ light + moisture', df, groups=df['site'])
result = model.fit()

print(result.summary())
```
